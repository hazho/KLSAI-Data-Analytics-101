```python
rows = []
for i in range (1,13):
    item = export[export['Month'] == i]["Total Export (USD)"].idxmax()
    rows.append(item)
# using a for loop, obtain the row which has the maximum total export in the month 
```

```python
rows  # index number corresponding with the highest value in the month 
```

```python
max_permonth = export.loc[rows] # make new dataset with filtered rows
```

```python
max_permonth # view new dataset
```

```python
max_permonth = max_permonth.reset_index(drop=True) # index above is according to the previous dataset
                                                   # reset the dataset according to the current dataset
```

```python
max_permonth
```

```python
# Plotting graphs
```

```python
import matplotlib.pyplot as plt # import matplotlib as plt for plotting graphs 

```

```python
month = max_permonth["Month"].values # obtain values in month column
```

```python
month
```

```python
exp = max_permonth["Total Export (USD)"].values # obtain values from the Export column
```

```python
exp
```

```python
# Plot graph of month against export for Maximum Export Per Month according to country
plt.plot(month, exp, color='g')
plt.xlabel('Month')
plt.ylabel('Export (USD)')
plt.title('Maximum Export per Month')
plt.show()
```

```python
sum(export[export['Month'] == 1]["Total Export (USD)"]) #obtain sum of export for month 1
```

```python
tpm = []
for i in range (1,13):
    total_per_month = sum(export[export['Month'] == i]["Total Export (USD)"])
    tpm.append(total_per_month)
# for loop to obtain the amount of export for each month 
```

```python
tpm
```

```python
# Plot graph of Total Export per month throughout the year 
plt.plot(month, tpm, color='r')

plt.xlabel('Month')
plt.ylabel('Export (USD)')
plt.title('Total Export per Month')
plt.show()
```

```python
# Compare between maximum export per month and total export 
plt.plot(month, exp, color='g')
plt.plot(month, tpm, color='r')
plt.xlabel('Month')
plt.ylabel('Export (USD)')
plt.title('Total Export per Month')
plt.show()
```
